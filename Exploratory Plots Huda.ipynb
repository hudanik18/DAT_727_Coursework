```python
#import here
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
%matplotlib inline

# Increase default figure and font sizes for easier viewing.
plt.rcParams['figure.figsize'] = (8, 6)
plt.rcParams['font.size'] = 14
```

## Reading The Data

```python
col_names = ['sex', 'length', 'diameter', 'height', 'whole_weight', 'shucked_weight', 'viscera_weight', 'shell_weight', 'rings']
path = 'https://archive.ics.uci.edu/ml/machine-learning-databases/abalone/abalone.data'
abalone = pd.read_csv(path, sep=',', header = 0, names = col_names)
```

```python
abalone.head()
```

```python
abalone.groupby('rings').count()
```

## Barplots

#### Exercise 1 Barplot of Rings
Start by making a barplot of the `rings` column:  

- each unique value in rings should be a bar (a category)  
- the number of abalones in each category should be the length of the bar
- set the figure size to 12 by 8
- add a title

```python

ab_plot = abalone.groupby('rings').count().plot(kind = 'bar',
                figsize = (12, 8))

ab_plot.set_title('Abalone Ring Sizes',
            fontsize = 21,
            y = 1.05,
            x= 0.5);

ab_plot.legend(loc=1);

ab_plot.set_xlabel("Ring Size",
             fontsize = 16)


# y-axis labels
ab_plot.set_ylabel('Number of Abalones',
             fontsize = 16);
```

#### Exercise 2 Dodged Barplot of Rings and Sex

## Scatter Plots

#### Exercise 3: Scatter Plot
- create a scatter plot of `length` on the x-azis & `diameter` on the y-axis
- set the figure size to 10 by 10 
- add labels to the axes
- give the plot a title

```python
ab_plot = abalone.plot(kind = 'scatter',
           x = 'length',
           y = 'diameter',
           color = 'purple',
              figsize = (10,10));
ab_plot.set_title('Abalone Length v Diameter',
            fontsize = 21,
            y = 1.05,
            x= 0.5);
ab_plot.set_xlabel("Length",
             fontsize = 16)


# y-axis labels
ab_plot.set_ylabel('Diameter',
             fontsize = 16);
```

#### Exercise 4 : Your Own Plotting Function
- write a function with takes three arguments
 - dataframe
 - x_variable
 - y_variable 
- makes a 12 by 12 scatter plot of the variable
- labels the x and y axes with the variable names
- set a custom color of your choice

Make four separate plots using numerical variables from the Abalone set using your function

```python
def scatterplot(df, x_axis, y_axis):
    ab_plot = df.plot(kind = 'scatter',
        x = x_axis,
        y = y_axis,
        color = 'orange',
        figsize = (12,12));
    ab_plot.set_title('Abalone ' +str(x_axis) + ' v ' + str(y_axis),
            fontsize = 21,
            y = 1.05,
            x= 0.5);
    ab_plot.set_xlabel(x_axis,
             fontsize = 16)
    ab_plot.set_ylabel(y_axis, fontsize = 16)

```

```python
# plot 1
scatterplot(abalone, 'length', 'sex')
```

```python
# plot 2
scatterplot(abalone, 'length', 'diameter')
```

```python
# plot 3
scatterplot(abalone, 'sex', 'whole_weight')
```

```python
# plot 4
scatterplot(abalone, 'whole_weight', 'height')
```
